## Range slider

```python
import dash
import dash_html_components as html
import dash_core_components as dcc

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div([
    html.H1("Select Data with RangeSlider", style={"text-align":"center","color":"blue"}),
    dcc.Graph(id="graph-with-rangeslider"),
    dcc.RangeSlider(
        id = "year-rangeslider",
        min = df["year"].min(),
        max = df["year"].max(),
        value = [1952, 1957],  # 
        marks = {str(year):str(year) for year in df["year"].unique()},  # 
        step = None  # 
    ),
#     html.Div(id='output-container-range-slider')
])


@app.callback(
    Output("graph-with-rangeslider","figure"),  # 
#     Output('output-container-range-slider', 'children'),
    [Input("year-rangeslider","value")]
)


def update_figure(selected_year):  # 

    # concat
    filtered_df0 = df[df.year == selected_year[0]]
    filtered_df1 = df[df.year == selected_year[1]]
    filtered_df = pd.concat([filtered_df0,filtered_df1])
#     return px.scatter(filtered_df,x="gdpPercap",y="lifeExp",log_x=True,hover_data=["continent","country"],color="country")

    traces = []

    for i in filtered_df.continent.unique():   # continent 
        df_by_continent = filtered_df[filtered_df["continent"] == i]  # 
        traces.append(dict(
            x = df_by_continent["gdpPercap"],  # 
            y = df_by_continent["lifeExp"],
            text = df_by_continent["country"],  # 
            mode = "markers",  # 
            opacity = 0.7,  # 
            marker = {  # marker
                "size":15,  # 
                "line":{"width":0.7, "color":"white"}
            },
            name = i   # trace
        ))

    return {   # 
        "data":traces,  # 
        "layout":dict(
            xaxis = {"type":"log",  # xlog
                     "title":"GDP for Per Captial",
                    "range": [2.3, 4.8]},  # 
            yaxis = {"title":"Life Expectancy", "range":[20,90]},
            margin={'l': 40, 'b': 40, 't': 10, 'r': 10},   # 
            legend = {"x":0,"y":1},
            hovermode = "closest",  # 
            transition = {"duration":500},
        )
    }

if __name__ == '__main__':
    app.run_server()
```
